```python
import uuid

class Ledger:
    def __init__(self):
        self.store = {}

    def create(self, value, parents=None, operation=None):
        obj_id = str(uuid.uuid4())[:8]

        self.store[obj_id] = {
            "value": value,
            "parents": parents or [],
            "operation": operation
        }
        return obj_id

    def trace(self, obj_id):
        chain = []
        
        def dfs(node):
            data = self.store[node]
            chain.append((node, data["value"], data["operation"]))
            for p in data["parents"]:
                dfs(p)
        
        dfs(obj_id)
        return chain

    def show(self):
        for k, v in self.store.items():
            print(k, "→", v)


# ---- Logic Layer ----
ledger = Ledger()

x1 = ledger.create(5, operation="init")

x2_val = ledger.store[x1]["value"] * 2
x2 = ledger.create(x2_val, parents=[x1], operation="*2")

x3_val = ledger.store[x2]["value"] + 10
x3 = ledger.create(x3_val, parents=[x2], operation="+10")


print("Final Value:", ledger.store[x3]["value"])

print("\nTrace of x3:")
for step in ledger.trace(x3):
    print(step)
```
