```python
import json
import os


MEMORY_FILE = "nlp_memory.json"

if os.path.exists(MEMORY_FILE):
    with open(MEMORY_FILE, "r") as f:
        memory = json.load(f)
else:
    memory = {"positive": [], "negative": [], "neutral": []}
def detect_emotion(text):
    positive_words = ["happy", "love", "great", "good", "awesome", "win"]
    negative_words = ["sad", "bad", "angry", "hate", "pain", "loss"]

    score = 0
    for word in text.lower().split():
        if word in positive_words:
            score += 1
        elif word in negative_words:
            score -= 1

    if score > 0:
        return "positive"
    elif score < 0:
        return "negative"
    else:
        return "neutral"
def detect_intent(text):
    text = text.lower()
    if "?" in text:
        return "question"
    elif any(word in text for word in ["want", "need", "wish"]):
        return "desire"
    elif any(word in text for word in ["feel", "thinking", "alone"]):
        return "emotion"
    else:
        return "statement"
def learn(text, emotion):
    if text not in memory[emotion]:
        memory[emotion].append(text)
def generate_response(emotion, intent):
    if emotion == "positive":
        return "You're in a strong state. Use it before it fades."
    elif emotion == "negative":
        return "Something’s off. Don’t ignore it — understand it."
    elif intent == "question":
        return "You’re searching. Good — that’s where growth starts."
    else:
        return "Interesting… say more."
def save_memory():
    with open(MEMORY_FILE, "w") as f:
        json.dump(memory, f, indent=4)
print("🧠 NLP Engine Started (type 'exit' to stop)\n")

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        save_memory()
        print("Memory saved. Goodbye.")
        break

    emotion = detect_emotion(user_input)
    intent = detect_intent(user_input)

    learn(user_input, emotion)

    response = generate_response(emotion, intent)

    print(f"AI ({emotion} | {intent}): {response}\n")
```
