```python
import random
templates = {
    "inquiry": [
        "Thank you for reaching out. I'd be happy to help you with this.",
        "Appreciate your message. Let me check this and get back to you shortly.",
        "Thanks for your interest. Here are the details you requested:"
    ],
    "delay": [
        "Apologies for the delay. I appreciate your patience.",
        "Sorry for the late response. Here's the update:",
        "Thanks for waiting. Here's the current status:"
    ],
    "closing": [
        "Let me know if you have any further questions.",
        "Looking forward to your response.",
        "Happy to assist further if needed."
    ]
}

def generate_reply(user_message):
    msg = user_message.lower()

    if "price" in msg or "cost" in msg:
        category = "inquiry"
    elif "waiting" in msg or "delay" in msg:
        category = "delay"
    else:
        category = "inquiry"

    reply = random.choice(templates[category])
    closing = random.choice(templates["closing"])

    return f"{reply}\n\n{closing}"
print("📧 Email Reply Generator\n")

while True:
    msg = input("Client message: ")
    if msg.lower() == "exit":
        break

    response = generate_reply(msg)
    print("\nSuggested Reply:\n")
    print(response)
    print("-" * 40)
```
