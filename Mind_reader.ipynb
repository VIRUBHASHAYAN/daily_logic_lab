```python
import random

class MindReader:
    def __init__(self):
        self.low = 1
        self.high = 100
        self.history = []

    def smart_guess(self):
        
        bias = random.choice([-2, -1, 0, 1, 2])
        guess = (self.low + self.high) // 2 + bias

        
        guess = max(self.low, min(self.high, guess))
        return guess

    def update_range(self, guess, feedback):
        self.history.append((guess, feedback))

        if feedback == "higher":
            self.low = guess + 1
        elif feedback == "lower":
            self.high = guess - 1

    def play(self):
        print("Think of a number between 1 and 100 😏")
        print("I will read your mind...")

        attempts = 0

        while True:
            guess = self.smart_guess()
            attempts += 1

            print(f"My guess: {guess}")
            feedback = input("Type 'higher', 'lower', or 'correct': ").lower()

            if feedback == "correct":
                print(f"😈 I knew it! Found in {attempts} moves.")
                break
            elif feedback in ["higher", "lower"]:
                self.update_range(guess, feedback)
            else:
                print("Invalid input... try again.")


game = MindReader()
game.play()
```
