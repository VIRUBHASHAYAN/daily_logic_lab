## Distance Based Recommendation

### Euculidion Approach

```python
from sklearn.metrics import pairwise_distances
```

```python
pivot_table_Db_Distance = pd.pivot_table(data =Movie_Db,values='rating',index='userId',columns='movie',).fillna(0)
pivot_table_Db_Distance.columns  = Movie_Db['movie'].unique()
pivot_table_Db_Distance.index    = Movie_Db['userId'].unique()
pivot_table_Db_Distance.iloc[:10,:10]
#pivot_table_Db.corr().round(1)
```

```python
distance_corr= pd.DataFrame(pairwise_distances(pivot_table_Db_Distance,metric='euclidean'))
distance_corr.columns  = Movie_Db['userId'].unique()
distance_corr.index    = Movie_Db['userId'].unique()
distance_corr.iloc[:10,:10].round(2)
```

```python
distance_corr= pd.DataFrame(pairwise_distances(pivot_table_Db_Distance,metric='euclidean'))
distance_corr.columns  = Movie_Db['userId'].unique()
distance_corr.index    = Movie_Db['userId'].unique()
distance_corr.iloc[:10,:10].round(2)
```
