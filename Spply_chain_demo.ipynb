```python
import random
import time

class SupplyChain:
    def __init__(self):
        self.inventory = 100
        self.reorder_level = 30
        self.reorder_amount = 70
        self.orders_completed = 0
        self.orders_failed = 0

    def generate_order(self):
        return random.randint(5, 25)

    def process_order(self, order):
        print(f"\n🛒 New Order: {order} units")

        if self.inventory >= order:
            self.inventory -= order
            self.orders_completed += 1
            print(f"✅ Order fulfilled. Remaining stock: {self.inventory}")
        else:
            self.orders_failed += 1
            print(f"❌ Not enough stock! Order failed.")

    def check_restock(self):
        if self.inventory <= self.reorder_level:
            print("⚠️ Low inventory! Restocking...")
            delay = random.randint(1, 3)
            time.sleep(delay)  # simulate delay
            self.inventory += self.reorder_amount
            print(f"📦 Restocked! New inventory: {self.inventory}")

    def simulate(self, rounds=10):
        for day in range(1, rounds + 1):
            print(f"\n📅 Day {day}")
            order = self.generate_order()
            self.process_order(order)
            self.check_restock()

        self.report()

    def report(self):
        print("\n📊 Final Report:")
        print(f"✔️ Completed Orders: {self.orders_completed}")
        print(f"❌ Failed Orders: {self.orders_failed}")
        print(f"📦 Final Inventory: {self.inventory}")

if __name__ == "__main__":
    sc = SupplyChain()
    sc.simulate(15)
```
