```python
def max_difference(arr):

    min_value = arr[0]
    max_diff = 0

    for i in range(1, len(arr)):

        diff = arr[i] - min_value

        max_diff = max(max_diff, diff)

        min_value = min(min_value, arr[i])

    return max_diff

arr = [7,1,5,3,6,4]

print(max_difference(arr))
```
