```python
import pandas as pd
insurance_db=pd.read_csv('Insurance.csv')
insurance_db.head()
```

##### Group_by function

#### Pull up insurance charges based on region 

```python
insurance_db.groupby(by='region')["charges"].mean().round(1).sort_values(ascending=False)
```

#### Pull up insurance charges based on region and sex

```python
insurance_db.groupby(['region','sex'])["charges"].mean().round(1).sort_values()
```

#### 2.Pivot_table function 

```python
pd.pivot_table(data = insurance_db,values='charges',index='region',columns='sex',aggfunc='max').round()
```

#### Aggregate function

```python
insurance_db.groupby(by='region')["charges"].agg(func=['min','max','sum']).round()
```

```python
def pof (x) :
    if x>20:
        return "Eligible"
    else:
        return "Not eligible"
```

```python
insurance_db['eligibility_criteria'] = insurance_db['age'].apply(pof)
insurance_db
```

```python
insurance_db['eligibility_criteria_2'] = insurance_db['age'].apply(lambda x:"Eligible" if x>20 else "Not Eligible")
insurance_db
```
