```python
import random
import time

class HumanGlitchText:
    def __init__(self, glitch_prob=0.2, delay_range=(0.02, 0.08)):
        self.glitch_prob = glitch_prob
        self.delay_range = delay_range

    def _glitch_char(self, char):
        glitches = [
            lambda c: c + c,                    # repeat
            lambda c: c.upper(),               # sudden caps
            lambda c: '',                      # skip
            lambda c: random.choice('!?@#'),   # random emotion
            lambda c: c + random.choice('...') # hesit
        ]
        return random.choice(glitches)(char)

    def type(self, text):
        output = ""
        for char in text:
            if random.random() < self.glitch_prob:
                char = self._glitch_char(char)

            print(char, end='', flush=True)
            output += char
            time.sleep(random.uniform(*self.delay_range))

        print()
        return output


# Example usage
if __name__ == "__main__":
    typer = HumanGlitchText(glitch_prob=0.3)

    text = "I didn't mean to say that... it just happened."
    typer.type(text)
```
