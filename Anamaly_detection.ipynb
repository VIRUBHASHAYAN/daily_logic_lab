```python
import numpy as np
from sklearn.ensemble import IsolationForest

# Simulated normal data (e.g., user behavior: time spent, clicks)
np.random.seed(42)
normal_data = np.random.normal(loc=50, scale=5, size=(100, 2))

# Add anomalies (fraud / unusual behavior)
anomalies = np.array([[100, 100], [5, 120], [200, 1]])

# Combine dataset
X = np.vstack([normal_data, anomalies])

# Train model
model = IsolationForest(contamination=0.05)
model.fit(X)

# Predict (-1 = anomaly, 1 = normal)
predictions = model.predict(X)

# Separate results
normal_points = X[predictions == 1]
anomaly_points = X[predictions == -1]

print("Normal Data Points:\n", normal_points[:5])
print("\nAnomalies Detected:\n", anomaly_points)
```
