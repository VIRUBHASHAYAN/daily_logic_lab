```python
class HabitPredictor:
    def __init__(self):
        self.history = []

    def add_day(self, completed):
        self.history.append(1 if completed else 0)

    def completion_rate(self):
        if len(self.history) == 0:
            return 0
        return sum(self.history) / len(self.history)

    def current_streak(self):
        streak = 0
        for day in reversed(self.history):
            if day == 1:
                streak += 1
            else:
                break
        return streak

    def predict_tomorrow(self):
        rate = self.completion_rate()
        streak = self.current_streak()

        score = (rate * 0.6) + (min(streak, 10) / 10 * 0.4)

        if score > 0.7:
            return "🔥 High chance you'll complete tomorrow"
        elif score > 0.4:
            return "⚠️ 50-50 chance — discipline needed"
        else:
            return "❌ Low chance — fix your consistency"

```

```python
tracker = HabitPredictor()

data = list(map(int, input("Enter last 10 days: ").split()))

for day in data:
    tracker.add_day(day)

print("Completion Rate:", tracker.completion_rate())
print("Current Streak:", tracker.current_streak())
print("Prediction:", tracker.predict_tomorrow())
```
