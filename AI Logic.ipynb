```python
import random

class SimpleAI:
    def __init__(self):
        # Step 1: Knowledge base (training data simulation)
        self.knowledge = {
            "hello": ["Hi there!", "Hey!", "Hello 👋"],
            "how are you": ["I'm fine.", "Doing great!", "All good!"],
            "ai": ["AI means machines thinking like humans.",
                   "AI learns patterns from data.",
                   "AI predicts the next best output."]
        }

    def process_input(self, user_input):
        # Step 2: Clean + understand input
        return user_input.lower().strip()

    def find_intent(self, processed_input):
        # Step 3: Match input with known patterns
        for key in self.knowledge:
            if key in processed_input:
                return key
        return "unknown"

    def generate_output(self, intent):
        # Step 4: Generate response
        if intent in self.knowledge:
            return random.choice(self.knowledge[intent])
        else:
            return "I don't understand, but I'm learning..."

    def run(self, user_input):
        processed = self.process_input(user_input)
        intent = self.find_intent(processed)
        response = self.generate_output(intent)

        # Show full pipeline
        return {
            "input": user_input,
            "processed": processed,
            "intent": intent,
            "output": response
        }


# 🔥 Demo
ai = SimpleAI()

while True:
    user = input("You: ")
    result = ai.run(user)

    print("\n--- AI THINKING ---")
    for k, v in result.items():
        print(f"{k}: {v}")
    print("-------------------\n")
```
