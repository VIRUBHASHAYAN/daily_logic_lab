```python
def first_unique_char(s):
    char_info = {}

    for i, ch in enumerate(s):
        if ch in char_info:
            char_info[ch] = -1   
        else:
            char_info[ch] = i    

    min_index = float('inf')
    result = None

    for ch, idx in char_info.items():
        if idx != -1 and idx < min_index:
            min_index = idx
            result = ch

    return result


# Example
text = input("Enter the string : ")
print(first_unique_char(text))  
```
