```python
def compress(text):
    words = text.split()
    freq = {}
    
   
    for word in words:
        freq[word] = freq.get(word, 0) + 1

    
    repeated = {w: i for i, (w, c) in enumerate(freq.items()) if c > 1}

    compressed = []
    dictionary = {}

    for word in words:
        if word in repeated:
            code = f"${repeated[word]}"
            compressed.append(code)
            dictionary[code] = word
        else:
            compressed.append(word)

    return " ".join(compressed), dictionary

def decompress(compressed_text, dictionary):
    words = compressed_text.split()
    result = []

    for word in words:
        if word in dictionary:
            result.append(dictionary[word])
        else:
            result.append(word)

    return " ".join(result)
text = input("Enter a sentence:\n")

compressed, dictionary = compress(text)

print("\nCompressed:", compressed)
print("Dictionary:", dictionary)

original = decompress(compressed, dictionary)
print("\nDecompressed:", original)
```
