### Example for Inheritance

```python
class Person:
    def __init__(self, name):
        self.name = name
    def introduce(self):
        print("Hello, my name is", self.name)
class Student(Person):
    def study(self):
        print(self.name, "is studying")
class Teacher(Person):
    def teach(self):
        print(self.name, "is teaching")
s = Student("Virubha")
t = Teacher("Tr.silba")
s.introduce()
s.study()
t.introduce()
t.teach()
```

### Example for polymorphism

```python
class EmailNotification:
    def send(self):
        print("Sending email notification")
class SMSNotification:
    def send(self):
        print("Sending SMS notification")
class AppNotification:
    def send(self):
        print("Sending app push notification")
notifications = [EmailNotification(), SMSNotification(), AppNotification()]
for n in notifications:
    n.send()
```

### Example for Abstraction

```python
from abc import ABC, abstractmethod
class Camera(ABC):
    @abstractmethod
    def capture(self):
        pass
class PhoneCamera(Camera):
    def capture(self):
        print("Capturing photo with AI stabilization")
class SecurityCamera(Camera):
    def capture(self):
        print("Capturing surveillance footage")
cam1 = PhoneCamera()
cam2 = SecurityCamera()
cam1.capture()
cam2.capture()
```
