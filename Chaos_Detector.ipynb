```python
import os, math

def get_entropy(s):
    probs = [s.count(c)/len(s) for c in set(s)]
    return -sum(p * math.log2(p) for p in probs)

def scan_mood(folder="."):
    files = [f for f in os.listdir(folder) if os.path.isfile(f)]
    if not files: return "Empty"
    avg_entropy = sum(get_entropy(f) for f in files) / len(files)
    
    if avg_entropy > 3.5: return "🔥 Chaotic Energy Detected"
    if avg_entropy > 2.5: return "⚖️ Mild Turbulence"
    return "🧘 Zen Organization Mode"

print(f"Current folder mood: {scan_mood()}")
```
