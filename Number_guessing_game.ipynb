```python
print("welcome to number guessing game!")
print("guess the number between 1-100")
key=34

for i in range(1,4):
   user=int(input("guess the secret number"))
   if user <0 or user >100:
    print("enter the valid number:(")
   elif user>key:
     print("guess is high")
   elif user<key:
      print("guess is low")
   else :
    print("congrats!! your guess is correct")
    print(f"you guess the number at your {i} attempt")
    break
if key!=user:
    print("sorry trial ended")
```

# while loop concept

```python
import random
print("welcome to number guessing game!")
print("guess the number between 1-100")
key=random.randint(1,100)

while True:
   user=int(input("guess the secret number"))
   if user <0 or user >100:
    print("enter the valid number:(")
   elif user>key:
     print("guess is high")
   elif user<key:
      print("guess is low")
   else :
    print("congrats!! your guess is correct")
    break
```
