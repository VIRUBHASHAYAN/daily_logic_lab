```python
from collections import defaultdict

def one_char_diff(a, b):

    if len(a) != len(b):
        return False

    diff = 0

    for x, y in zip(a, b):

        if x != y:
            diff += 1

        if diff > 1:
            return False

    return diff == 1


def longest_chain(words):

    graph = defaultdict(list)

    for i in range(len(words)):

        for j in range(i + 1, len(words)):

            if one_char_diff(words[i], words[j]):

                graph[words[i]].append(words[j])
                graph[words[j]].append(words[i])

    memo = {}

    def dfs(node, visited):

        key = (node, tuple(sorted(visited)))

        if key in memo:
            return memo[key]

        best = 1

        for nei in graph[node]:

            if nei not in visited:

                best = max(
                    best,
                    1 + dfs(nei, visited | {nei})
                )

        memo[key] = best

        return best

    ans = 0

    for word in words:

        ans = max(ans, dfs(word, {word}))

    return ans

words = ["cold", "cord", "card", "ward", "warm", "worm"]

print(longest_chain(words))
```
