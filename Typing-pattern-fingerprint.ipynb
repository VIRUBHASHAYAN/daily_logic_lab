### This code analyzes how a user types a sentence and creates a unique fingerprint based on:

* vowel vs consonant flow
* word length pattern
* repetition style

```python
def typing_fingerprint(text):
    vowels = "aeiouAEIOU"
    
    vowel_count = 0
    consonant_count = 0
    pattern = []
    
    words = text.split()
    
    for word in words:
        v = sum(1 for char in word if char in vowels)
        c = sum(1 for char in word if char.isalpha() and char not in vowels)
        
        vowel_count += v
        consonant_count += c
        
        pattern.append(len(word))
    
    fingerprint = f"V{vowel_count}-C{consonant_count}-P{pattern}"
    
    return fingerprint
text = input("Enter your sentence: ")
print("Your typing fingerprint:", typing_fingerprint(text))
```
