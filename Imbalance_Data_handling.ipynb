```python
import pandas as pd
from sklearn.datasets import load_breast_cancer,load_diabetes
from matplotlib import pyplot as plt
```

```python
Dataset = load_breast_cancer()
```

```python
brest_cancer_db=pd.DataFrame(data = Dataset.data , columns=Dataset.feature_names)
```

```python
brest_cancer_db
```

```python
print(Dataset.DESCR)
```

```python
brest_cancer_db['Target'] = Dataset.target
```

```python
brest_cancer_db
```

```python
brest_cancer_db['Target'].value_counts()
```

```python
plt.pie(x = brest_cancer_db['Target'].value_counts() ,labels = ['Benign','Malignant'],autopct = "%1.1f%%")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
 x = brest_cancer_db.drop("Target",axis = 1 )
```

```python
y = brest_cancer_db['Target']
```

```python
%%time
from sklearn.preprocessing import StandardScaler
std_scalar= StandardScaler()
Scallar_X=pd.DataFrame(std_scalar.fit_transform(X = x),columns = Dataset.feature_names)
```

```python
Scallar_X.mean(),Scallar_X.std()
```

```python
x_train, x_test, y_train, y_test= train_test_split(Scallar_X,y,test_size = 0.2,random_state = 12,stratify = y)
```

```python
 Model = DecisionTreeClassifier(class_weight={0:2,1:1})

 Model.fit(x_train,y_train)
```

```python
y_pred_train = Model.predict(x_train)
```

```python
y_pred_test = Model.predict(x_test)
```

```python
print(classification_report(y_train,y_pred_train))
```

```python
print(classification_report(y_test,y_pred_test))
```

```python
print(confusion_matrix(y_train,y_pred_train))
```

```python
print(confusion_matrix(y_test,y_pred_test))
```
