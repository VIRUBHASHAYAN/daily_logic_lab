```python


class Request:
    def __init__(self, url, method="GET", data=None):
        self.url = url
        self.method = method
        self.data = data


class Response:
    def __init__(self, content, status=200):
        self.content = content
        self.status = status


class Database:
    def __init__(self):
        self.users = {"admin": "1234"}

    def get_user(self, username):
        return self.users.get(username)


class Server:
    def __init__(self):
        self.db = Database()

    def handle_request(self, request):
        print(f"Incoming request → {request.url}")

        # Routing (deciding what to do based on URL)
        if request.url == "/":
            return self.home()

        elif request.url == "/login" and request.method == "POST":
            return self.login(request.data)

        else:
            return Response("404 Not Found", 404)

    def home(self):
        return Response("Welcome to the website")

    def login(self, data):
        username = data.get("username")
        password = data.get("password")

        stored_password = self.db.get_user(username)

        if stored_password == password:
            return Response("Login Successful")
        else:
            return Response("Invalid Credentials", 401)


class Client:
    def send_request(self, server, url, method="GET", data=None):
        request = Request(url, method, data)
        response = server.handle_request(request)
        self.render(response)

    def render(self, response):
        print(f"Response [{response.status}] → {response.content}")


# ---- Simulation ----
server = Server()
client = Client()

# User visits homepage
client.send_request(server, "/")

# User tries login
client.send_request(server, "/login", method="POST",
                    data={"username": "admin", "password": "1234"})

# Wrong login
client.send_request(server, "/login", method="POST",
                    data={"username": "admin", "password": "wrong"})
```
