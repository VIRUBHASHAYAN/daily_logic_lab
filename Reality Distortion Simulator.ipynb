```python
import random

class RealityEngine:
    def __init__(self):
        self.timeline = []
        self.energy = 50 
        self.focus = 50    

    def take_decision(self, decision):
        """
        decision: 'work', 'waste', 'learn', 'rest'
        """
        impact = {
            "work":    ( +10, +5 ),
            "waste":   ( -15, -10 ),
            "learn":   ( +5, +15 ),
            "rest":    ( +8, +8 )
        }

        if decision not in impact:
            print("Invalid decision")
            return

        e_change, f_change = impact[decision]

    
        noise = random.randint(-5, 5)

        self.energy += e_change + noise
        self.focus += f_change + noise

        # clamp values
        self.energy = max(0, min(100, self.energy))
        self.focus = max(0, min(100, self.focus))

        self.timeline.append((decision, self.energy, self.focus))

    def predict_future(self):
        """
        Predict outcome based on current state
        """
        score = (self.energy * 0.5) + (self.focus * 0.5)

        if score > 75:
            return "High Performance Path 🚀"
        elif score > 50:
            return "Stable but Average Path ⚖️"
        elif score > 25:
            return "Struggling Path ⚠️"
        else:
            return "Collapse Path 💀"

    def show_timeline(self):
        print("\n--- Timeline ---")
        for i, step in enumerate(self.timeline, 1):
            print(f"Day {i}: {step[0]} → Energy:{step[1]} Focus:{step[2]}")

        print("\nFinal State Prediction:", self.predict_future())



engine = RealityEngine()

days = int(input("How many days to simulate: "))

for i in range(days):
    d = input(f"Day {i+1} decision (work/waste/learn/rest): ")
    engine.take_decision(d)

engine.show_timeline()
```
